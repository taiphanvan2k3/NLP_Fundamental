```python
import urllib.request as urllib2
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = urllib2.urlopen("https://vtv.vn/cong-nghe.htm")
html_doc = response.read()
```

```python
soup = BeautifulSoup(html_doc, "html.parser")
pretty_html = soup.prettify()
```

```python
print(pretty_html)
```

```python
news_list = soup.find_all("div", class_="list_news")

data = []

# Lặp qua từng div.list_news
for news in news_list:
    li_tags = news.find_all("li", class_="tlitem")
    # Lấy tất cả các thẻ <li> trong div này
    for li in li_tags:
        title = li.h4.a.text.strip()
        link = li.h4.a["href"]
        thumbnail = li.img["src"]
        description = li.find("p", class_="sapo").text.strip()

        data.append(
            {
                "Title": title,
                "Link": link,
                "Thumbnail": thumbnail,
                "Description": description,
            }
        )
```

```python
df = pd.DataFrame(data)
df
```

### Lưu vào CSV

```python
df.to_csv("vtv_news.csv", index=False)
```
